## 1. Dataset description

This dataset is a **cleaned version** of the original Mushroom Dataset for **binary classification**, sourced from the UCI Library. It has been preprocessed using techniques like **modal imputation**, **one-hot encoding**, **z-score normalization**, and **feature selection**.  

The dataset was obtained from [Kaggle - Mushroom Dataset by Prisha Sawhney](https://www.kaggle.com/datasets/prishasawhney/mushroom-dataset).  


---

### Features:  
1. **Cap Diameter**: Numerical feature representing the diameter of the mushroom cap.  
2. **Cap Shape**: Shape of the cap.  
3. **Gill Attachment**: How the gills are attached to the stem.  
4. **Gill Color**: Color of the gills.  
5. **Stem Height**: Height of the mushroom stem (numerical).  
6. **Stem Width**: Width of the mushroom stem (numerical).  
7. **Stem Color**: Color of the stem.  
8. **Season**: The season when the mushroom is found.  

---

### Target Variable:  
- **Target Class**: Binary classification label:  
   - `0` ‚Üí Edible  
   - `1` ‚Üí Poisonous  

---

### Summary:  
The dataset contains **9 columns** and has already undergone data cleaning. This makes it a suitable choice for training classification models without dealing with missing values or extensive preprocessing.

### 3. Data splitting

We will divide the dataset as follows:  
- **60%** for training data  
- **20%** for testing data  
- **20%** for validation data  
This split ensures a balanced distribution for model training, testing, and evaluation.  


```python
y = mushroom_df['class']
x = mushroom_df.drop('class', axis=1)

Xtrain, Xtest, ytrain, ytest = train_test_split(x, y, test_size=0.4, random_state=random_seed)

display(Xtrain.shape)
display(ytrain.shape)
display(Xtest.shape)
display(ytest.shape)
```

```python
Xval, Xtest, yval, ytest = train_test_split(Xtest, ytest, test_size=0.5, random_state=random_seed)

display(Xval.shape)
display(yval.shape)
display(Xtest.shape)
display(ytest.shape)
```

We will create a graph to illustrate how the data is divided into training, validation, and test sets. This will help ensure the splits are clear and properly balanced.  

```python
def plot_data_distribution(Xtrain, Xval, Xtest):
    sizes = [len(Xtrain), len(Xval), len(Xtest)]
    labels = ['Training Set', 'Validation Set', 'Test Set']
    percentages = [size / sum(sizes) * 100 for size in sizes]

    plt.figure(figsize=(8, 6))
    plt.bar(labels, percentages, color=['skyblue', 'lightgreen', 'lightcoral'], edgecolor='black')
    plt.title("Data Distribution Across Training, Validation, and Test Sets")
    plt.ylabel("Percentage")
    plt.ylim(0, 100)
    
    for i, v in enumerate(percentages):
        plt.text(i, v + 1, f"{v:.2f}%", ha='center', fontsize=10)

    plt.grid(linewidth = .4, alpha = .4)
    plt.show()
```

```python
plot_data_distribution(Xtrain, Xval, Xtest)
```

#### Conclusion of Data Preparation  
The data has been cleaned, missing values handled, categorical features encoded, and the dataset split into training, testing, and validation sets. We are now ready to move on to model training.  


## 3. Model

We need to prepare data subsets for KNN. For this, we perform min-max normalization.

```python
scaler = MinMaxScaler()
scaler.fit(Xtrain)

Xtrain_scaled = scaler.transform(Xtrain)
Xval_scaled = scaler.transform(Xval)
Xtest_scaled = scaler.transform(Xtest)
```

We iterate over all hyperparameter combinations for KNN, train the model, and store accuracy for training and validation sets; then, we select the parameters with the highest validation accuracy.

```python
param_grid = {
    'n_neighbors': range(5, 20),  
    'weights': ['uniform', 'distance'],
    'metric': ['minkowski', 'manhattan']  
}

params_comb = ParameterGrid(param_grid)
```

```python
train_accuracy = []
val_accuracy = []
train_f1 = []
val_f1 = []

for params in params_comb:
    knn = KNeighborsClassifier(**params)  
    knn.fit(Xtrain_scaled, ytrain)  

    ytrain_pred = knn.predict(Xtrain_scaled)
    yval_pred = knn.predict(Xval_scaled)

    train_accuracy.append(accuracy_score(ytrain, ytrain_pred))
    val_accuracy.append(accuracy_score(yval, yval_pred))
    train_f1.append(f1_score(ytrain, ytrain_pred))
    val_f1.append(f1_score(yval, yval_pred))

results = pd.DataFrame(params_comb)
results['train_accuracy'] = train_accuracy
results['val_accuracy'] = val_accuracy
results['train_f1'] = train_f1
results['val_f1'] = val_f1

plt.style.use('default')  
```

```python
def plot_knn_results(results):
    fig, ax = plt.subplots(1, 2, figsize=(16, 8))
    
    ax[0].plot(range(len(results)), results['train_accuracy'], 'o-', label='Train Accuracy', color='blue')
    ax[0].plot(range(len(results)), results['val_accuracy'], 'o-', label='Validation Accuracy', color='orange')
    ax[0].set_title('Accuracy for k-NN Hyperparameters')
    ax[0].set_xlabel('Parameter Combination Index')
    ax[0].set_ylabel('Accuracy')
    ax[0].legend()
    ax[0].grid()  

    ax[1].plot(range(len(results)), results['train_f1'], 'o-', label='Train F1-Score', color='green')
    ax[1].plot(range(len(results)), results['val_f1'], 'o-', label='Validation F1-Score', color='purple')
    ax[1].set_title('F1-Score for k-NN Hyperparameters')
    ax[1].set_xlabel('Parameter Combination Index')
    ax[1].set_ylabel('F1-Score')
    ax[1].legend()
    ax[1].grid()

    plt.tight_layout()
    plt.show()
```

```python
plot_knn_results(results)
```

# **Accuracy (left):**

**Train Accuracy (blue)**: Close to 1.0 for most hyperparameters, indicating overfitting as the model memorizes training data.

**Validation Accuracy (orange):** Much lower than training accuracy, showing the model struggles to generalize. Variations suggest sensitivity to hyperparameters like n_neighbors.


# **F1-Score (right):**

**Train F1-Score (green):** 
Also near 1.0, reinforcing overfitting.

**Validation F1-Score (purple):** Lower and fluctuates similarly to validation accuracy, reflecting how the model's balance of precision and recall varies.


# **Insights:**
Overfitting is evident, with high training metrics but lower validation performance.
Hyperparameters like n_neighbors and metric strongly influence validation results.
Improving generalization may require regularization or fine-tuning hyperparameters to reduce the performance gap.

```python
best_index = results['val_accuracy'].idxmax()  
best_params = params_comb[best_index]

knn = KNeighborsClassifier(**best_params)
knn.fit(Xtrain_scaled, ytrain)

train_accuracy = metrics.accuracy_score(ytrain, knn.predict(Xtrain_scaled))
val_accuracy = metrics.accuracy_score(yval, knn.predict(Xval_scaled))
test_accuracy = metrics.accuracy_score(ytest, knn.predict(Xtest_scaled))

print('accuracy score (train): {0:.6f}'.format(train_accuracy))
print('accuracy score (validation): {0:.6f}'.format(val_accuracy))
print('accuracy score (test): {0:.6f}'.format(test_accuracy))
```

```python
ConfusionMatrixDisplay.from_estimator(knn, Xtest_scaled, ytest)
plt.show()
```

The model performs well in predicting both classes. It correctly classified majority of instances of class 0 and of class 1. There is a relatively small number of false positives and false negatives. The model has great accuracy 

```python
y_test_prob = knn.predict_proba(Xtest_scaled)[:, 1]

fpr, tpr, roc_thresholds = roc_curve(ytest, y_test_prob)
roc_auc = auc(fpr, tpr)

precision, recall, pr_thresholds = precision_recall_curve(ytest, y_test_prob)
average_precision = average_precision_score(ytest, y_test_prob)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

axes[0].plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
axes[0].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
axes[0].set_xlabel('False Positive Rate')
axes[0].set_ylabel('True Positive Rate')
axes[0].set_title('Receiver Operating Characteristic (ROC) Curve')
axes[0].legend(loc='lower right')
axes[0].grid()

axes[1].plot(recall, precision, color='blue', lw=2, label=f'PR Curve (AP = {average_precision:.2f})')
axes[1].set_xlabel('Recall')
axes[1].set_ylabel('Precision')
axes[1].set_title('Precision-Recall Curve')
axes[1].legend(loc='lower left')
axes[1].grid()

plt.tight_layout()
plt.show()
```

These plots visualize the performance of the k-NN model using **ROC** and **Precision-Recall (PR) curves**:

**ROC Curve (left):**

The curve shows the trade-off between True Positive Rate (sensitivity) and False Positive Rate for various thresholds.
The Area Under the Curve (AUC) is 0.99, indicating excellent model performance, as the curve is close to the top-left corner.

**PR Curve (right):**

This curve focuses on the relationship between precision (positive predictive value) and recall (sensitivity).
The Average Precision (AP) score is 0.99, further confirming that the model performs well in distinguishing between classes, even with imbalanced data.

**Insights:**
Both metrics demonstrate near-perfect model performance, with high precision, recall, and AUC scores.
The model is well-calibrated, minimizing false positives and negatives effectively.
